```python
def mi_decorador(funcion_original):
    def envoltorio():
        print("1. Antes de ejecutar la función...")
        funcion_original()
        print("2. Después de ejecutar la función.")
    return envoltorio

def saludar():
    print("   ¡Hola mundo!")

# Aplicación manual
saludar_decorado = mi_decorador(saludar)
saludar_decorado()
```

```python
@mi_decorador
def despedirse():
    print("   Adiós, me voy.")

despedirse()
```
